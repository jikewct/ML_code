```python
import torch
import numpy as np
import scipy
import matplotlib.pyplot as plt
import cv2
import imageio
```

```python
print(imageio.help())
```

```python
image_cat = cv2.imread("../../data/images/cat.jpg")
image_cat = cv2.cvtColor(image_cat, cv2.COLOR_BGR2RGB)
img = cv2.resize(image_cat, (400,400))
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sigma = 0.8
noise = np.random.randn(*(img_gray.shape)) * sigma
perturbed_img = img_gray / 255 + noise
fig = plt.figure()
#print(plt.cm.get_cmap().name)
ax1 = fig.add_subplot(1,4,1 )
ax1.imshow(img, plt.cm.viridis)
ax2 = fig.add_subplot(1,4,2)
ax2.imshow(img_gray, plt.cm.gray)
ax3 = fig.add_subplot(1,4,3)
ax3.imshow(noise, plt.cm.gray)
ax4 = fig.add_subplot(1,4,4)
ax4.imshow(perturbed_img, plt.cm.gray)
plt.show()
print(image_cat.shape,img.shape, img_gray.shape)
cv2.imwrite("../../data/images/gray_cat.jpg", img_gray)
imageio.imwrite("../../data/images/gray_cat1.jpg", img_gray.astype(np.uint8))
```
